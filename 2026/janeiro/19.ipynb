In Quantum Chromodynamics with light quarks, the classical Lagrangian exhibits an approximate chiral symmetry 
$SU(N_f)_L \times SU(N_f)_R$, corresponding to independent rotations of left- and right-handed quark fields. 
In the QCD vacuum, however, this symmetry is **spontaneously broken** by the formation of a nonzero chiral condensate 
$\langle \bar q q \rangle \neq 0$, which dynamically couples left- and right-handed components. Importantly, the vacuum 
does not select one chiral sector over the other; instead, it *locks* them together, reducing the symmetry to the diagonal 
vector subgroup $SU(N_f)_V$. This breaking reflects a choice of orientation in chiral space rather than an explicit 
preference for left or right chirality, and it leads to the emergence of (pseudo-)Goldstone bosons, such as the pions, 
which encode the broken symmetry structure of the QCD vacuum. **Lattice QCD** é uma abordagem numérica não perturbativa para estudar a Cromodinâmica Quântica (QCD), usada quando o acoplamento forte é grande e métodos perturbativos falham. A ideia central é discretizar o espaço-tempo contínuo em uma rede (lattice) finita, onde os quarks vivem nos sítios da rede e os glúons nos elos entre os sítios, preservando a simetria de gauge. As integrais funcionais da QCD são então avaliadas por métodos de Monte Carlo estatístico, permitindo calcular propriedades fundamentais diretamente a partir da teoria, como massas de hádrons, constantes de decaimento, equação de estado da matéria fortemente interagente e a temperatura de transição para o plasma de quarks e glúons. Em particular, lattice QCD é essencial para estudar confinamento, restauração da simetria quiral e transições de fase em QCD, fornecendo previsões quantitativas que servem como referência teórica para experimentos de íons pesados.
