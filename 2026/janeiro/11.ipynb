Gaussian Processes inferem uma distribuição posterior sobre o espaço de todas as funções compatíveis com os dados e com as correlações impostas pelo kernel.

Espaço de funções possíveis: todas as funções que são compatíveis com uma estrutura Gaussiana de correlação k(xi,xj).

A Gaussian Process is fully specified by a mean function and a covariance function.
