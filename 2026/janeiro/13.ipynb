um enunciado de **conserva√ß√£o por transporte no espa√ßo de fase** (6D).

> Ou seja: o lado esquerdo da Boltzmann √© literalmente ‚ÄúLiouville para $f$‚Äù.

---

### 2) Evid√™ncia 2: o termo de colis√£o √© constru√≠do para respeitar invariantes microsc√≥picos

O termo de colis√£o n√£o pode ser qualquer coisa. Ele √© obrigado a respeitar o fato de que colis√µes microsc√≥picas (bin√°rias, el√°sticas) conservam:

- n√∫mero de part√≠culas,
- momento linear,
- energia.

Isso se traduz em tr√™s identidades (os chamados *colisional invariants*):

$$
\int \left(\frac{\partial f}{\partial t}\right)_{\text{col}}\, d^3v = 0,
$$

$$
\int m\mathbf{v}\,\left(\frac{\partial f}{\partial t}\right)_{\text{col}}\, d^3v = 0,
$$

$$
\int \frac{1}{2}m v^2\,\left(\frac{\partial f}{\partial t}\right)_{\text{col}}\, d^3v = 0.
$$

Essas equa√ß√µes s√£o uma evid√™ncia forte de que o termo de colis√£o:
- n√£o √© um ‚Äútermo arbitr√°rio‚Äù,
- √© um operador com simetrias e conserva√ß√µes herdadas da microdin√¢mica.

> Isso √© exatamente o tipo de ‚Äúrestri√ß√£o estrutural‚Äù que aparece quando voc√™ reduz Liouville com hip√≥teses f√≠sicas.

---

### 3) Evid√™ncia 3: o termo de colis√£o produz a seta do tempo (H-teorema)

Com Liouville, a entropia de Gibbs √© conservada (reversibilidade).  
Com Boltzmann, existe o H-teorema, que essencialmente diz:

$$
\frac{d}{dt}\left[
- \int f\ln f\, d^3x\,d^3v
\right] \ge 0.
$$

O ponto-chave √©:

- o lado esquerdo (Liouville/streaming) **n√£o produz** crescimento de entropia,
- o lado direito (colis√µes + hip√≥tese estat√≠stica) √© o que **quebra** a reversibilidade.

> Ou seja: a irreversibilidade n√£o est√° ‚Äúna mec√¢nica‚Äù, mas no fechamento estat√≠stico que cria o termo de colis√£o.

---

## Conclus√£o visual (no mesmo esp√≠rito Euler ‚Üí Navier‚ÄìStokes)

- **Liouville (exato, $6N$D):**
$$
\frac{\partial \rho_N}{\partial t} + \{\rho_N,H_N\} = 0
$$

- **Boltzmann (reduzido, 6D):**
$$

+ \mathbf{v}\cdot\nabla_{\mathbf{x}} f
+ \mathbf{F}\cdot\nabla_{\mathbf{v}} f
=
\textcolor{#f1c40f}{\left(\frac{\partial f}{\partial t}\right)_{\text{col}}}
$$

A evid√™ncia de transi√ß√£o √© que:
1. sem o termo amarelo, Boltzmann vira Liouville reduzido;
2. o termo amarelo √© rigidamente constrangido por conserva√ß√µes microsc√≥picas;
3. o termo amarelo √© onde nasce a seta do tempo (H-teorema).

## Mini-deriva√ß√£o BBGKY em ~10 linhas 

**(1)** Considere $N$ part√≠culas cl√°ssicas com estados $\Gamma_i=(\mathbf{x}_i,\mathbf{v}_i)$ e densidade completa  
$$
\rho_N(\Gamma_1,\dots,\Gamma_N,t).
$$

**(2)** A din√¢mica exata satisfaz a **equa√ß√£o de Liouville**:
$$
\partial_t \rho_N + \sum_{i=1}^N \mathbf{v}_i\!\cdot\!\nabla_{\mathbf{x}_i}\rho_N
+ \sum_{i=1}^N \frac{\mathbf{F}_i}{m}\!\cdot\!\nabla_{\mathbf{v}_i}\rho_N = 0,
$$
onde $\mathbf{F}_i=\sum_{j\neq i}\mathbf{F}_{ij}$ √© a for√ßa total sobre $i$ (intera√ß√µes de pares).

**(3)** Defina a distribui√ß√£o de 1 part√≠cula (reduzida):
$$
f_1(\Gamma_1,t)=N\!\int \rho_N(\Gamma_1,\dots,\Gamma_N,t)\, d\Gamma_2\cdots d\Gamma_N.
$$

**(4)** Integre a equa√ß√£o de Liouville em $d\Gamma_2\cdots d\Gamma_N$. Os termos com derivadas em $\mathbf{x}_i,\mathbf{v}_i$ para $i\ge2$ viram termos de contorno e somem (assumindo decaimento/condi√ß√µes apropriadas).

**(5)** O que sobra d√°, para $f_1$, um termo de transporte + um termo de intera√ß√£o:
$$
\partial_t f_1 + \mathbf{v}_1\!\cdot\!\nabla_{\mathbf{x}_1} f_1
+ \frac{\mathbf{F}_{\text{ext}}(\mathbf{x}_1)}{m}\!\cdot\!\nabla_{\mathbf{v}_1} f_1
= - \int \frac{\mathbf{F}_{12}}{m}\!\cdot\!\nabla_{\mathbf{v}_1}\, f_2(\Gamma_1,\Gamma_2,t)\, d\Gamma_2.
$$

**(6)** Aqui aparece a distribui√ß√£o de 2 part√≠culas:
$$
f_2(\Gamma_1,\Gamma_2,t)=N(N-1)\!\int \rho_N\, d\Gamma_3\cdots d\Gamma_N.
$$

**(7)** Ou seja: a equa√ß√£o para $f_1$ **n√£o fecha** ‚Äî ela depende de $f_2$.  
Isso j√° √© o ‚ÄúDNA‚Äù BBGKY:  
$$
f_1 \leftarrow f_2 \leftarrow f_3 \leftarrow \cdots
$$

**(8)** Repetindo o procedimento para $f_2$, voc√™ obt√©m:
$$
\partial_t f_2 + (\text{termos de transporte em }1,2) = \int (\text{for√ßas})\, f_3\, d\Gamma_3,
$$
isto √©: $f_2$ depende de $f_3$, etc.

**(9)** **Fechamento (hip√≥tese do caos molecular):**
$$
f_2(\Gamma_1,\Gamma_2,t)\approx f_1(\Gamma_1,t)\,f_1(\Gamma_2,t)
\quad (\text{antes da colis√£o}).
$$

**(10)** Substituindo esse fechamento na eq. de $f_1$, o termo de intera√ß√£o vira um operador efetivo que √© justamente o **termo de colis√£o**:
$$
\partial_t f + \mathbf{v}\!\cdot\!\nabla_{\mathbf{x}} f + \frac{\mathbf{F}_{\text{ext}}}{m}\!\cdot\!\nabla_{\mathbf{v}} f
=
\left(\frac{\partial f}{\partial t}\right)_{\text{col}}.
$$

> Moral em uma frase: **Liouville (exato) ‚Üí integra√ß√£o reduzida d√° $BBGKY$ ‚Üí fechamento $f_2\simeq f_1f_1$ cria o termo de colis√£o de Boltzmann.**
