### Small Systems 

Colisoes de ions pequenos desafiam o entendimento, já que esperava-se que a menor quantidade de material e o menor tempo causasse sistemas em regime perturbativo, alpha << 1, em que não há tempo suficiente para termalização, mas o que se observa são os mesmos padrões observados em colisões de íons pesados. [Progress and Challenges in Small Systems](https://arxiv.org/pdf/2401.09208)

Os dados observados, coletividade e anisotropias sugere uma fase hidrodinâmica. Mas sabemos que essa fase é limitada pelo número de Raynolds !
