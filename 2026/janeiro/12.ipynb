## Origem cinemática da expansão em harmônicos $ v_n $

**Analisando apenas a cinemática**, após o *freeze-out* temos um **conjunto de partículas detectadas**. Nesse estágio, o sistema já não evolui dinamicamente: não há forças, nem equações de movimento, apenas um estado final congelado acessível experimentalmente.

Fixando o momento transversal $ p_T $, podemos medir a distribuição angular

$$
\frac{dN}{d\phi}(p_T),
$$

que representa o **número de partículas com momento transversal $ p_T $** emitidas na **direção azimutal $ \phi $**.

A variável $ \phi $ é um ângulo, portanto satisfaz a identificação geométrica

$$
\phi \equiv \phi + 2\pi.
$$

Segue que a função $ \frac{dN}{d\phi}(p_T) $ é uma **função periódica de período $ 2\pi $**.

---

### Expansão em série de Fourier

Toda função periódica definida em um círculo pode ser expandida em **série de Fourier**. Assim, **necessariamente existe** uma expansão do tipo

$$
\frac{dN}{d\phi}(p_T)
=
a_0(p_T)
+
\sum_{n=1}^{\infty}
\left[
a_n(p_T)\cos(n\phi)
+
b_n(p_T)\sin(n\phi)
\right].
$$

Essa expansão é um **resultado puramente matemático**, decorrente apenas da periodicidade da variável angular $ \phi $.

---

### Escolha de um ângulo de referência

Em cada evento, escolhe-se um **ângulo de referência** $ \Psi_n $, correspondente a uma **rotação global no plano transversal**. Essa escolha é cinemática e não altera a distribuição física, apenas redefine o eixo angular.

Com essa rotação, o par de coeficientes trigonométricos pode ser reescrito como

$$
a_n(p_T)\cos(n\phi) + b_n(p_T)\sin(n\phi)
=
A_n(p_T)\cos\big(n(\phi - \Psi_n)\big),
$$

onde $ A_n(p_T) $ e $ \Psi_n $ são **apenas reparametrizações trigonométricas**, sem novo conteúdo físico.

---

### Normalização da distribuição

Define-se o número total de partículas naquele momento transversal $ p_T $ como

$$
N(p_T)
=
\int_0^{2\pi}
d\phi\,
\frac{dN}{d\phi}(p_T).
$$

O termo médio da série de Fourier é então dado por

$$
a_0(p_T)
=
\frac{N(p_T)}{2\pi}.
$$

Essa definição garante que a distribuição angular esteja **corretamente normalizada**.

---

### Definição dos coeficientes $ v_n $

Por convenção, definem-se os coeficientes harmônicos como

$$
v_n(p_T)
\equiv
\left\langle
\cos\big(n(\phi - \Psi_n)\big)
\right\rangle,
$$

ou, de forma explícita,

$$
v_n(p_T)
=
\frac{
\int_0^{2\pi}
d\phi\,
\cos\big(n(\phi - \Psi_n)\big)\,
\frac{dN}{d\phi}(p_T)
}{
\int_0^{2\pi}
d\phi\,
\frac{dN}{d\phi}(p_T)
}.
$$

Esses coeficientes correspondem a **projeções da distribuição angular** sobre a base harmônica natural do círculo.

---

### Forma final da expansão

Substituindo todas as definições anteriores, obtém-se a forma final da expansão:

$$
\frac{dN}{d\phi}(p_T)
=
\frac{N(p_T)}{2\pi}
\left[
1
+
2
\sum_{n=1}^{\infty}
v_n(p_T)
\cos\big(n(\phi - \Psi_n)\big)
\right].
$$

Essa expressão é **cinemática, estatística e geométrica**, não envolvendo qualquer equação de movimento.

---

### O modo isotrópico $ v_0 $

Podemos, então, identificar o **modo harmônico de ordem zero** como o termo médio da expansão. O coeficiente correspondente é

$$
v_0(p_T)
\equiv
\frac{1}{2\pi}
\int_0^{2\pi}
d\phi\,
\frac{dN}{d\phi}(p_T)
=
\frac{N(p_T)}{2\pi}.
$$

O termo $ v_0(p_T) $ representa a **contribuição isotrópica média** da distribuição angular, servindo como **normalização** sobre a qual os coeficientes $ v_n(p_T) $, com $ n \ge 1 $, quantificam as anisotropias do estado final.

Sobre $ \mathscr{v_0(p_T)} $.


Fixando o momento transversal $p_T$, a distribuição azimutal de partículas pode ser expandida em uma série de Fourier como


$$
\frac{dN}{d\phi}(p_T)
=
\frac{N(p_T)}{2\pi}
\left[
1 + 2\sum_{n=1}^{\infty} v_n(p_T)\cos\big(n(\phi-\Psi_n)\big)
\right].
$$


O coeficiente $v_0(p_T)$ surge ao projetarmos essa distribuição sobre o modo constante (harmônico de ordem zero), isto é,


$$
v_0(p_T)
\;\propto\;
\frac{1}{2\pi}\int_0^{2\pi}\frac{dN}{d\phi}(p_T)\,d\phi.
$$


Como


$$
\int_0^{2\pi}\cos\big(n(\phi-\Psi_n)\big)\,d\phi = 0 \quad \text{para } n\ge1,
$$


resta apenas o termo isotrópico, de modo que $v_0(p_T)$ representa a contribuição média angular — o <em>fundo isotrópico</em> — da produção de partículas naquele $p_T$.
Assim, $v_0(p_T)$ não codifica anisotropia, mas estabelece a normalização física sobre a qual os coeficientes $v_n(p_T)$, $n\ge1$, quantificam as quebras de isotropia da colisão.
