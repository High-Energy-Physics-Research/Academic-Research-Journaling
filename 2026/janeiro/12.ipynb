<p><em>Sobre</em> $ \mathscr{v_0(p_T)} $.</p>

<p>
Fixando o momento transversal $p_T$, a distribuição azimutal de partículas pode ser expandida em uma série de Fourier como
</p>

$$
\frac{dN}{d\phi}(p_T)
=
\frac{N(p_T)}{2\pi}
\left[
1 + 2\sum_{n=1}^{\infty} v_n(p_T)\cos\big(n(\phi-\Psi_n)\big)
\right].
$$

<p>
O coeficiente $v_0(p_T)$ surge ao projetarmos essa distribuição sobre o modo constante (harmônico de ordem zero), isto é,
</p>

$$
v_0(p_T)
\;\propto\;
\frac{1}{2\pi}\int_0^{2\pi}\frac{dN}{d\phi}(p_T)\,d\phi.
$$

<p>
Como
</p>

$$
\int_0^{2\pi}\cos\big(n(\phi-\Psi_n)\big)\,d\phi = 0 \quad \text{para } n\ge1,
$$

<p>
resta apenas o termo isotrópico, de modo que $v_0(p_T)$ representa a contribuição média angular — o <em>fundo isotrópico</em> — da produção de partículas naquele $p_T$.
Assim, $v_0(p_T)$ não codifica anisotropia, mas estabelece a normalização física sobre a qual os coeficientes $v_n(p_T)$, $n\ge1$, quantificam as quebras de isotropia da colisão.
</p>
