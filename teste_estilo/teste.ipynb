Em fluidos relativísticos, partimos da conservação local do tensor energia–momento,
expressa pela equação fundamental
$$
\partial_\mu T^{\mu\nu} = 0.
$$

Para um fluido ideal, o tensor energia–momento assume a forma
$$
T^{\mu\nu} = (\varepsilon + p)\, u^\mu u^\nu - p\, g^{\mu\nu},
$$
onde $\varepsilon$ é a densidade de energia, $p$ a pressão e $u^\mu$ o quadrivetor
velocidade, normalizado por $u^\mu u_\mu = 1$.

Projetando a equação de conservação ao longo de $u_\nu$, obtemos
$$
u_\nu \partial_\mu T^{\mu\nu}
= u^\mu \partial_\mu \varepsilon
+ (\varepsilon + p)\, \partial_\mu u^\mu = 0.
$$

Definindo a derivada convectiva ao longo da linha de mundo do fluido como
$d/d\tau = u^\mu \partial_\mu$, a equação acima pode ser reescrita como
$$
\frac{d\varepsilon}{d\tau}
= -(\varepsilon + p)\, \nabla_\mu u^\mu.
$$

Em uma expansão longitudinal simples, como no caso de Bjorken,
o divergente do campo de velocidades é dado por $\nabla_\mu u^\mu = 1/\tau$,
de modo que a equação de conservação de energia se reduz a
$$
\frac{d\varepsilon}{d\tau}
= -\frac{\varepsilon + p}{\tau}.
$$

Assumindo uma equação de estado do tipo $p = c_s^2\, \varepsilon$,
essa equação diferencial pode ser integrada diretamente,
$$
\int_{\varepsilon_0}^{\varepsilon(\tau)}
\frac{d\varepsilon'}{\varepsilon' (1 + c_s^2)}
=
-\int_{\tau_0}^{\tau} \frac{d\tau'}{\tau'}.
$$

O resultado da integração fornece a dependência explícita da densidade de energia
com o tempo próprio,
$$
\varepsilon(\tau)
= \varepsilon_0
\left( \frac{\tau_0}{\tau} \right)^{1 + c_s^2}.
$$

Essa expressão deixa claro como a expansão do fluido governa o resfriamento
do sistema em hidrodinâmica relativística.
